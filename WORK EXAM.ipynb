```python
'1、Convert the app sizes to a number'
```

```python
df["Size"] = df["Size"].apply(lambda x: x.replace(',', ''))
powers = {'k': 10 ** 3, 'M': 10 ** 6, '+': 10 ** 0}
# add some more to powers as necessary

def f(s):
    if 'Varies with device' in s:q=np.nan
    else:
        try:
            power = s[-1]
            return float(s[:-1]) * powers[power]
        except TypeError:
            return s
        
df["Size"] = df["Size"].apply(f)
print(df["Size"].head())
```

```python
'2、Convert the number of installs to a number'
```

```python
print(df["Installs"].drop_duplicates())
df["Installs"] = df["Installs"].apply(lambda x: x.replace(',', '').replace('+',''))
def getnum2(s):
    if 'Free' in s:
        q=np.nan
    else:
        return int(s)

df["Installs"] = df["Installs"].apply(getnum2)
print(df["Installs"].head())
```

```python
def ban_name(Senti):
    name_ban = ["positive","Neutral","Negative","I","nan","app","very beautiful","it"," can","much","bank","this","user","calendar","weapons","Android",'E',"Relax","dark","good","keep","AFK"]
    if any(x in Senti for x in ["40%"]):
        pass
    else:
        if any(x in Senti for x in name_ban):
            Senti = np.nan
    return Senti

```

```python
df4["Senti"]=df4['Senti'].astype(str)
df4["Senti"] = df4["Senti"].apply(ban_name)
df4.dropna(how="any")
```
